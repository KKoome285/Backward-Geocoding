```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geo_colab")
```

```python
latitude = -1.3793
longitude = 36.6515

location = geolocator.reverse((latitude, longitude), exactly_one=True)
print(location.address)
```

```python
coords = [(-1.3793, 36.6515), (-1.4342, 36.6481), (-1.4746, 36.9632)]  # Ngong, Kiserian, Kitengela
addresses = []

for lat, lon in coords:
    loc = geolocator.reverse((lat, lon), exactly_one=True)
    if loc:
        addresses.append({"latitude": lat, "longitude": lon, "address": loc.address})

import pandas as pd
df = pd.DataFrame(addresses)
df
```
